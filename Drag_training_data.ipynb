```python
from docx import Document
import pandas as pd

doc = Document('./Desktop/GA_V1.docx')

generations = []
records = []
best_scores = []
average_scores = []

for para in doc.paragraphs:
    line = para.text
    if line.startswith('generation'):
        parts = line.split(',')
        generations.append(int(parts[0].split(':')[1].strip()))
        records.append(int(parts[1].split(':')[1].strip()))
        best_scores.append(int(parts[2].split(':')[1].strip()))
        average_scores.append(float(parts[3].split(':')[1].strip()))

data = pd.DataFrame({
    'Generation': generations,
    'Record': records,
    'Best Score': best_scores,
    'Average Score': average_scores
})

data
```

```python
data.to_csv('GA_Training_process_1')
```

```python
from docx import Document
import pandas as pd

doc = Document('./Desktop/GA_V2.docx')

generations = []
records = []
best_scores = []
average_scores = []

for para in doc.paragraphs:
    line = para.text
    if line.startswith('generation'):
        parts = line.split(',')
        generations.append(int(parts[0].split(':')[1].strip()))
        records.append(int(parts[1].split(':')[1].strip()))
        best_scores.append(int(parts[2].split(':')[1].strip()))
        average_scores.append(float(parts[3].split(':')[1].strip()))

data = pd.DataFrame({
    'Generation': generations,
    'Record': records,
    'Best Score': best_scores,
    'Average Score': average_scores
})

data
```

```python
data.to_csv('GA_Training_process_2')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(data['Generation'], data['Best Score'])
plt.title('Generation vs Best Score')
plt.xlabel('Generation')
plt.ylabel('Best Score')
plt.grid(True)
plt.show()
```

```python
plt.rcParams.update({'font.size': 14})

plt.figure(figsize=(20, 6))

plt.subplot(1, 2, 1)
plt.plot(data['Generation'], data['Best Score'])
plt.title('Generation vs Best Score', fontsize=20)
plt.xlabel('Generation', fontsize=14)
plt.ylabel('Best Score', fontsize=14)
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(data['Generation'], data['Average Score'],color='orange')
plt.title('Generation vs Average Score', fontsize=20)
plt.xlabel('Generation', fontsize=14)
plt.ylabel('Average Score', fontsize=14)
plt.grid(True)

plt.show()
```
